```python
'''
    DBSCAN bases its clusters off of high and low density areas
    It is typically robust to noise and outliers
    - For our uses we were unable to find clusters
'''
```

```python
# imports
import pandas as pd
from sklearn.preprocessing import StandardScaler
from kneed import KneeLocator
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.impute import SimpleImputer
from sklearn.datasets import make_blobs

```

```python
# prepare data for clustering
df = pd.read_csv("data/pbd_outliers_removed.csv")
#cols = [
#    "video_pause_4",
#    "video_play_4",
#    "video_watch_4",
#    "video_completed_4"
#]

cols = [
 "video_pause_1",
 "video_completed_1",
 "video_pause_4",
 "video_play_1",
 "video_watch_2",
 "video_watch_3",
 "video_seek_3",
 "video_play_2",
 "video_seek_4",
 "video_watch_1",
 "video_completed_2",
 "video_pause_2",
 "video_play_4",
 "video_play_3",
 "video_pause_3",
 "video_seek_1",
 "video_watch_4",
 "video_seek_2",
 "video_caption_1",
 "video_completed_3",
 "video_caption_2",
 "video_completed_4"
]


trimmed_df = df[cols]
#scores = df["ALG1_EOC_SCALE_SCORE_22"]

trimmed_df
```

```python
# scales data
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(trimmed_df), columns=trimmed_df.columns)

df_scaled
```

```python
# remove NaN
# this imputer completes missing values with the mean
imputer = SimpleImputer(strategy="mean")
df_scaled = imputer.fit_transform(df_scaled)
```

```python
# dimensionality scaling
'''
    Visualizes high-dimensional data by giving each data point a location
    in a two or three-dimensional map. 
'''
# synthetic data for testing
#df_scaled, _ = make_blobs(n_samples=300, centers=3, random_state=0)

tsne = TSNE(n_components=2, random_state = 0)
df_tsne = pd.DataFrame(tsne.fit_transform(df_scaled), columns=["TSNE1", "TSNE2"])
```

```python

db = DBSCAN(eps=3, min_samples=10).fit(df_tsne)
labels = db.labels_

n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print(f"Estimated number of clusters: {n_clusters}")
print(f"Estimated number of noise points: {n_noise_}")
```

```python
core_samples_mask = np.zeros_like(labels, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True

df_tsne['Cluster'] = labels
plt.scatter(df_tsne["TSNE1"], df_tsne["TSNE2"], c=df_tsne['Cluster'], cmap='viridis', marker='o', s=50)
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
